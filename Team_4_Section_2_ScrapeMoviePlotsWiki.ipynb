```python
#using Beautiful Soup and wikipedia library , extract all the plots of american movies released in year 2010-17
import wikipedia
import requests
from bs4 import BeautifulSoup
import time
import numpy as np
links=[]
movieName=[]
plot=[]
years=[]
#2011,2018
for year in range(2010,2018):
    html = requests.get('https://en.wikipedia.org/wiki/List_of_American_films_of_'+ str(year))
    
    #turn the HTML into a beautiful soup text object
    b = BeautifulSoup(html.text, 'lxml')
    # create an mpty list where those links will go.
    for i in b.find_all(name ='i')[3:-1]:
        for link in i.find_all('a', href=True):
            links.append(link.get('href'))
            movieName.append(link.get_text())
            print(str(year) + link.get_text())
            years.append(year)
            try:
                plot.append(wikipedia.WikipediaPage(title = link.get('href').split("/")[-1]).section('Plot').replace('\n','').replace("\'",""))
            except:
                plot.append(np.NaN)
```

```python
#create data frame and print to csv output
import pandas as pd
MoviePlot = pd.DataFrame(
    {'year': years,
     'movie': movieName,
     'link': links,
     'plot': plot,
    })
```

```python
# Exporting the plot data out to csv
MoviePlot.to_csv('MoviePlot2010to2017.csv',index=False)
```
